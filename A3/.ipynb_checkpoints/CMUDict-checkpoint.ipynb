```lua
function tail(list)
    return { select(3, unpack(list)) }
end

local file = io.open("cmudicttest.txt")
cmudict = {}
for line in file:lines() do
    local entry = line:split(" ")
    local word = entry[1]
    local phonemes = tail(entry)
    cmudict[word] = phonemes
end
print(cmudict["AABERG"])
```
